```python
import pyautogui
import win32gui

# Encuentra la ventana activa por su título
def find_window(title):
    def callback(handle, data):
        if win32gui.IsWindowVisible(handle) and title in win32gui.GetWindowText(handle):
            data.append(handle)
        return True
    handles = []
    win32gui.EnumWindows(callback, handles)
    return handles[0]

# Cambia 'Título de la ventana' por el título de la ventana donde quieres hacer clic y scroll
window_title = "Microsoft Power BI - Google Chrome"


# Encontrar todas las celdas de la tabla
cells = driver.find_elements(By.XPATH, "//div[@role='gridcell']")

# Obtener el número de columnas de la tabla
numero_de_columnas = len(cells) // len(driver.find_elements(By.XPATH, "//div[@role='row']"))

# Iterar sobre las celdas y extraer el texto
fila_actual = []
for i, cell in enumerate(cells):
    texto_celda = cell.text
    fila_actual.append(texto_celda)
    # Agregar la fila a la tabla cuando se alcance el final de una fila
    if (i + 1) % numero_de_columnas == 0:
        data.append(fila_actual)
        fila_actual = []

for i in range(310):
    
    # Encuentra y activa la ventana
    window_handle = find_window(window_title)
    win32gui.SetForegroundWindow(window_handle)

    # Obtiene las dimensiones de la ventana
    left, top, right, bottom = win32gui.GetWindowRect(window_handle)
    width = right - left
    height = bottom - top

    # Calcula las coordenadas del centro de la ventana
    center_x = left + width // 2
    center_y = top + height // 2

    # Da clic en el centro de la ventana
    pyautogui.click(center_x, center_y)

    # Hacer scroll hacia abajo 10 veces
    for _ in range(3):
        pyautogui.press("pagedown")
        
        
    # Encontrar todas las celdas de la tabla
    cells = driver.find_elements(By.XPATH, "//div[@role='gridcell']")

    # Obtener el número de columnas de la tabla
    numero_de_columnas = len(cells) // len(driver.find_elements(By.XPATH, "//div[@role='row']"))

    # Iterar sobre las celdas y extraer el texto
    fila_actual = []
    for i, cell in enumerate(cells):
        texto_celda = cell.text
        fila_actual.append(texto_celda)
        # Agregar la fila a la tabla cuando se alcance el final de una fila
        if (i + 1) % numero_de_columnas == 0:
            data.append(fila_actual)
            fila_actual = []

lista_aplanada = [item for sublist in data for item in sublist]
```

```python
# Lista de sub listas
sub_listas = []
sub_lista = []

# Recorrer la lista original
for elemento in lista_aplanada:
    # Si encontramos 'Seleccionar fila', agregamos la sub lista actual a la lista de sub listas
    if elemento == 'Seleccionar fila':
        if sub_lista:
            sub_listas.append(sub_lista)
            sub_lista = []
    # Agregamos el elemento a la sub lista actual
    sub_lista.append(elemento)

# Agregamos la última sub lista a la lista de sub listas
if sub_lista:
    sub_listas.append(sub_lista)

# Imprimimos la lista de sub listas
for sub_lista in sub_listas:
    print(sub_lista)
```

```python
data_2021 = pd.DataFrame(sub_listas, columns=["Selección", "Puesto", "Departamento", "Municipio",
                                               "Nombre del establecimiento", "Puntaje", "Estrellas"])
```
